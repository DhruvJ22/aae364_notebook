# Relative Stability

## Gain Margin

```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = control.tf([1, 0], [0, 1])
G = 0.1*(s-1)/(s+1)
control.nyquist(G, np.logspace(-2, 2));
plt.grid()
```

```python
control.bode(G, dB=True);
```

Z = N + P

Z = 0 + 0 = 0

```python
control.pole(G/(1+G))
```

If we increase gain too much, the closed loop system becomes unstable.

```python
s = control.tf([1, 0], [0, 1])
G2 = -2.5*(s-1)/(s+1)
control.nyquist(G2, np.logspace(-2, 2));
plt.grid()
```

```python
control.pole(G/(1+G))
```

We can compute the stability margins with the command:

```python
control.stability_margins(G)
```

We can also observe the stability margins from the bode plot.

```python
control.bode(G, dB=True, Hz=True, omega=np.logspace(-5, 5));
```

## Phase Margin

```python
G3 = 1/(s*(s+1))
control.nyquist(G3);
plt.grid()
```

```python
control.bode(G3, dB=True, Hz=True);
```

```python
control.stability_margins(G3)
```

```python
control.pole(G3/(1+G3))
```

```python
G4 = 10/(s*(s+10))
control.nyquist(G4);
plt.grid()
```

```python
control.bode(G4, dB=True, Hz=True);
```

```python
control.stability_margins(G4);
```

```python
D = -(s - 20)/(s + 20)
G = 1/s**2
H = (s + 1)
G5 = G*H*D
G5
```

```python
G5 = G*H*D
control.nyquist(G5);
plt.grid()
```

```python
control.pole(control.minreal(G5/(1 + G5)))
```

```python
control.stability_margins(G5)
```
